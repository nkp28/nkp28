```python
# For LVM files
# Basic Packages
import matplotlib.pyplot as plt
import numpy as np

# Module for reading lvm files
import lvm_read

filename = 'Cylinder_12OCT2021_Thurston_4.lvm'

lvm = lvm_read.read(filename, read_from_pickle=False)

fig, axis = plt.subplots(3)

# PLot each individually
x = []
for ix in range(len(lvm[0]['data'])):
    x = x + [lvm[0]['data'][ix][0]]
    
# Do table displacement
y1 = []
for ix in range(len(lvm[0]['data'])):
    y1 = y1 + [lvm[0]['data'][ix][1]]
axis[0].plot(x, y1)
axis[0].set_title("Load")
    
# Table Displacement
y2 = []
for ix in range(len(lvm[0]['data'])):
    y2 = y2 + [lvm[0]['data'][ix][2]]
axis[1].plot(x, y2)
axis[1].set_title("Table Displacement")
    
# Voltage Comment
y3 = []
for ix in range(len(lvm[0]['data'])):
    y3 = y3 + [lvm[0]['data'][ix][3]]
axis[2].plot(x, y3)
axis[2].set_title("Laser Extensomoter")

plt.xlabel('Time')
```
